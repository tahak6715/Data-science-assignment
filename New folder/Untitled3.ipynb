```python
import pandas as pd
data=pd.read_csv("pokemon.csv")
data.head(n=10)
```

```python
speed_mean = data["Speed"].mean()

def set_speed(val):
    if val < speed_mean:
        return "low speed"
    else:
        return "high speed"
```

```python
data["High_low_Speed"] = data["Speed"].apply(set_speed)
data
```

```python
HP_mean = data["HP"].mean()

def set_HP(val):
    if val < HP_mean:
        return "low HP"
    else:
        return "high HP"
```

```python
data["High_low_HP"] = data["HP"].apply(set_HP)
data
```
